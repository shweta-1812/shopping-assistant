# Prediction for most sold article

```python
import pandas as pd
```

```python
# Load filtered articles
df_filtered = pd.read_csv("../raw_data/articles_filtered.csv")

# Load filtered transactions (needed for sort_by_income)
df_trans_filtered = pd.read_csv("../raw_data/transactions_filtered.csv")

# Convert article_id to string
article_ids = df_filtered['article_id'].astype(str).tolist()
```

## Most sold by number of transactions

```python
# Load filtered transactions
df_trans = pd.read_csv("../raw_data/transactions_filtered.csv")

# Count number of times each article was sold
most_sold_count = df_trans['article_id'].value_counts()

# Top 1 most sold
top_article_id = most_sold_count.index[0]
top_count = most_sold_count.iloc[0]

print(f"Most sold article by number of transactions: {top_article_id} ({top_count} sales)")
```

## Most sold by total revenue (price sum)

```python
# Sum revenue per article
revenue_per_article = df_trans.groupby("article_id")["price"].sum()

# Top 1 by revenue
top_article_id = revenue_per_article.idxmax()
top_revenue = revenue_per_article.max()

print(f"Most sold article by revenue: {top_article_id} ({top_revenue})")
```

## Optional: Top 5

```python
# Top 5 articles by number of sales
top5_by_count = most_sold_count.head(5).to_frame("num_sales")

# Top 5 articles by total revenue
top5_by_revenue = revenue_per_article.sort_values(ascending=False).head(5).to_frame("total_revenue")

# Print both tables
print("Top 5 articles by number of sales:")
print(top5_by_count)

print("\nTop 5 articles by revenue:")
print(top5_by_revenue)
```

### Optional: round total revenue to 2 decimal places


```python
# Top 5 articles by number of sales
top5_by_count = most_sold_count.head(5).to_frame("num_sales")

# Top 5 articles by total revenue, rounded to 2 decimals
top5_by_revenue = (
    revenue_per_article.sort_values(ascending=False)
    .head(5)
    .round(2)  # rounds the values to 2 digits
    .to_frame("total_revenue")
)

# Print both tables
print("Top 5 articles by number of sales:")
print(top5_by_count)

print("\nTop 5 articles by revenue:")
print(top5_by_revenue)
```
